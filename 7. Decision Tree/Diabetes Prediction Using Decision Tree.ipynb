```python
from sklearn.datasets import load_diabetes
```

```python
ds= load_diabetes()
```

```python
ds
```

```python
print(ds['DESCR'])
```

```python
import pandas as pd

X= pd.DataFrame(ds.data, columns= ['age', 'sex', 'bmi', 'bp', 's1', 's2', 's3', 's4', 's5', 's6']) # Independent features
y= ds['target']
```

```python
X
```

```python
y
```

```python
# train test split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.3, random_state=10)
```

```python
X_train
```

```python
# correlation

X_train.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15,10))
sns.heatmap(X_train.corr(), annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor= DecisionTreeRegressor()
regressor.fit(X_train, y_train)
```

## Hyperparameter Tunning

```python
param= {
    'criterion': ['squared_error', 'friedman_mse', 'absolute_error'],
    'splitter':['best', 'random'],
    'max_depth': [1,2,3,4,5,10,20,25],
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
param
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
regressor= DecisionTreeRegressor()
```

```python
grid= GridSearchCV(regressor, param_grid= param, cv=5, scoring='neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings('ignore')

grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred= grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
print(r2_score(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
```

```python
selectmodel= DecisionTreeRegressor(criterion= 'squared_error', max_depth= 3,  max_features= 'auto', splitter= 'random')
```

```python
selectmodel.fit(X_train, y_train)
```

```python
# Visualize the decision tree

from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(selectmodel, filled=True)
```
